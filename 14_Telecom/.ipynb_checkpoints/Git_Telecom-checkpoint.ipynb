## Задача

Оператор связи «████» хочет бороться с оттоком клиентов. Для этого его сотрудники начнут предлагать промокоды и специальные условия всем, кто планирует отказаться от услуг связи. Чтобы заранее находить таких пользователей, «████» нужна модель, которая будет предсказывать, разорвёт ли абонент договор. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и услугах. **Задача — обучить на этих данных модель для прогноза оттока клиентов**.

## Описание услуг

**Оператор предоставляет два основных типа услуг**: 
- Стационарную телефонную связь. Телефон можно подключить к нескольким линиям одновременно.
- Интернет. Подключение может быть двух типов: через телефонную линию (`DSL`, от англ. digital subscriber line — «цифровая абонентская линия») или оптоволоконный кабель (`Fiber optic`).

**Также доступны такие услуги**:
- Интернет-безопасность: антивирус (`DeviceProtection`) и блокировка небезопасных сайтов (`OnlineSecurity`);
- Выделенная линия технической поддержки (`TechSupport`);
- Облачное хранилище файлов для резервного копирования данных (`OnlineBackup`);
- Стриминговое телевидение (`StreamingTV`) и каталог фильмов (`StreamingMovies`).

Клиенты могут платить за услуги каждый месяц или заключить договор на 1–2 года. Возможно оплатить счёт разными способами, а также получить электронный чек.

## Описание данных

Данные состоят из нескольких файлов, полученных из разных источников:
- `████.csv` — информация о договоре;
- `████.csv` — персональные данные клиента;
- `████.csv` — информация об интернет-услугах;
- `████.csv` — информация об услугах телефонии.

Информация о договорах актуальна на 1 февраля 2020 года.

#### Создание целевого признака

```python
████.head(3)
```

Логика такая: в чистом виде даты начала и окончания договора не нужны. Нам важно знать, сколько клиент с нами и закончил ли он сотрудничество или нет. Для этого мы:
- создадим столбец `Days`, где вычтем `BeginDate` из `EndDate` (в случае, если договор все еще действует, `EndDate` приравниваем к 1 февраля 2020 года)
- создадим столбец `Target`, куда поместим `0`, если клиент все еще с нами (`EndDate` = NaT), или `1`, если договор больше не действует

```python
#создадим отдельный столбец, чтобы потом не запутаься с таргетом
████['EndDate_1'] = ████['EndDate'].fillna(pd.to_datetime('2020-02-01'))

#cоздаем столбец Days, в котором считаем разницу
████['Days'] = (████['EndDate_1'] - ████['BeginDate']).dt.days

#создаем столбец Target
████['Target'] = ████['EndDate'].notna()
```

###### Распределение категориальных признаков

```python
#функция для отрисовки графиков
def make_graphic(n_cols, cols, data):
    n_cols = n_cols
    n_rows = math.ceil(len(cols) / n_cols)

    plt.figure(figsize=(14, 6 * n_rows))
    
    for i, col in enumerate(cols):
        plt.subplot(n_rows, n_cols, i + 1)
        ax = sns.countplot(data=data, x=col, palette='Blues')
        plt.title(col)
        plt.xticks(rotation=7)
        plt.ylabel('Количество')
        plt.xlabel('')
    
        total = len(data)
        
        for p in ax.patches:
            count = p.get_height()
            percent = 100 * count / total
            x = p.get_x() + p.get_width() / 2
            y = p.get_height()
            ax.text(x, y + total * 0.0005, f'{percent:.1f}%', ha='center', va='bottom', fontsize=8)

    plt.tight_layout()
    plt.show()
```

```python
cat_cols = ['Type', 'PaperlessBilling', 'PaymentMethod', 'Target']
make_graphic(2, cat_cols, ████)
```

## Шаг 5. Подготовка данных

Выполним подготовку данных для обучения модели. Разделим данные на две выборки, при обработке учтем, что нужно сделать OHE-кодирование для категориальных признаков и масштабирование для количественных. Для последнего используем `RobustScaler`, так как у нас есть выбросы в `TotalCharges`.

```python
RANDOM_STATE = 240325
TEST_SIZE = 0.25

X_train, X_test, y_train, y_test = train_test_split(
    all_data.drop(['Target'], axis=1),
    all_data['Target'],
    test_size = TEST_SIZE, 
    random_state = RANDOM_STATE,
    stratify = all_data['Target'])
```

```python
X_train.shape, X_test.shape
```

```python
# создаём списки с названиями признаков
ohe_columns = ['Type', 'PaperlessBilling', 'gender', 'SeniorCitizen', 'Partner',
               'Dependents', 'InternetService', 'OnlineSecurity', 'OnlineBackup',
               'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies',
               'MultipleLines', 'IsStable']
num_columns = ['MonthlyCharges', 'TotalCharges', 'Days']

# создаём пайплайн для подготовки признаков из списка ohe_columns
ohe_pipe = Pipeline(
    [
     ('ohe', OneHotEncoder(drop='first', handle_unknown='error', sparse_output=False))
    ]
    )

# создаём пайплайн для подготовки данных для логрега и рандомного леса
preprocessor_lr = ColumnTransformer(
    [('ohe', ohe_pipe, ohe_columns),
     ('num', RobustScaler(), num_columns)
    ], 
    remainder='drop'
)

# общий пайплайн
pipe_final = Pipeline([
    ('preprocessor', 'passthrough'),
    ('models','passthrough')
])
```

```python
# Прогоняем препроцессор
X_train_array = preprocessor_lr.fit_transform(X_train)
X_test_array = preprocessor_lr.transform(X_test)

# Получаем имена признаков из OHE
ohe_feature_names = preprocessor_lr.named_transformers_['ohe'].named_steps['ohe'].get_feature_names_out(input_features=ohe_columns)

# Финальный список имён (сначала OHE, потом числовые признаки)
all_feature_names = np.concatenate([ohe_feature_names, num_columns])

# Преобразуем в DataFrame
X_train_preprocessed_lr = pd.DataFrame(X_train_array, columns=all_feature_names, index=X_train.index)
X_test_preprocessed_lr = pd.DataFrame(X_test_array, columns=all_feature_names, index=X_test.index)
```

```python
#смотрим на итоги работы препроцессора для линейных моделей
display(X_train_preprocessed_lr.head(3))
display(X_test_preprocessed_lr.head(3))
```

## Шаг 6. Обучение моделей машинного обучения

Обучим три модели: логистическую регрессию, рандомный лес и катбуст.

```python
param_grid = [
    
    # словарь для модели LogisticRegression() 
    {
        'models': [LogisticRegression(random_state=RANDOM_STATE)],
        'models__C': [0.001, 0.01, 0.1, 1, 10, 100],
        'models__penalty': ['l2'],
        'models__solver': ['lbfgs', 'liblinear'],
        'models__max_iter': [100, 200, 500],
        'preprocessor': [preprocessor_lr]   
    },
    
    # словарь для модели RandomForestClassifier() 
    {
        'models': [RandomForestClassifier(random_state=RANDOM_STATE)],
        'models__n_estimators': [100, 200, 300],
        'models__max_depth': [None, 5, 10, 20],
        'models__max_features': ['sqrt', 'log2'],
        'models__bootstrap': [True, False],
        'preprocessor': [preprocessor_lr]
    }
]

    #словарь для модели CatBoostClassifier()
    
param_grid_cb = [
    {
        'models': [CatBoostClassifier(verbose=0, random_state=RANDOM_STATE)],
        #'models__iterations': [100, 300, 500],
        'models__learning_rate': [0.01, 0.05, 0.1],
        #'models__depth': [4, 6, 8, 10],
        #'models__l2_leaf_reg': [1, 3, 5, 7, 9],
        'models__bootstrap_type': ['Bayesian', 'Bernoulli'],
        'preprocessor': ['passthrough'],
        'models__cat_features': [ohe_columns]
    }
    
]
```

```python
joblib.parallel_backend('threading')
```

```python
#поиск для логрега и рандомного леса
random_search = RandomizedSearchCV(
    pipe_final, 
    param_grid,
    random_state=RANDOM_STATE,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1
)

#обучение
random_search.fit(X_train, y_train)
```

```python
#поиск для катбуста
random_search_cb = RandomizedSearchCV(
    pipe_final, 
    param_grid_cb,
    random_state=RANDOM_STATE,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1
)

#обучение
X_train[ohe_columns] = X_train[ohe_columns].astype('category')
random_search_cb.fit(X_train, y_train)
```

## Шаг 7. Выбор лучшей модели

Выберем лучшую модель и проверим её качество на тестовой выборке.

```python
if random_search.best_score_ > random_search_cb.best_score_:
    print('Лучшая модель и её параметры:\n\n', random_search.best_estimator_)
    print ('\n\nМетрика лучшей модели на кросс-валидации:', random_search.best_score_)
    print(random_search.best_estimator_.get_params())
else:
    print('Лучшая модель и её параметры:\n\n', random_search_cb.best_estimator_)
    print ('\n\nМетрика лучшей модели на кросс-валидации:', random_search_cb.best_score_)
    print(random_search_cb.best_estimator_.get_params())
```

Лучшая метрика на кросс-валидации получилась у CatBoost с `learning_rate` = 0.1 и `bootstrap_type` = Bernoulli. Посмотрим, насколько она опередила другие модели:

```python
print('Лучшая небустинговая модель:\n\n', random_search.best_estimator_)
print ('\n\nЕе оценка на кросс-валидации:', random_search.best_score_)
```

    - Распределение категориальных признаков сказало нам о том,
        1. Реже уходят те, кто платит ежемесячно
        2. Чаще уходят те, кто в отношениях и без детей
        3. Чаще уходят те, кто подключал `Fiber Optic`, `MultipleLines` и другие дополнительные услуги.
    - Распределение количественных признаков свидетельствует, что чаще прекращают договоры клиенты со средним "сроком жизни" и высокими ежемесячными платежами.
    - Корреляционный анализ показал много мультиколлинеарных признаков:
        1. Факт подключения интернета тесно связан с подключением сопутствующих услуг.
        2. Интернет-подключение сильно связано с `TotalCharges`.
        3. `StreamingTV` сильно связан со `StreamingMovies`.

4. **Разделили выборки, написали общий пайплайн и предобработали в нем данные**
    - Для обучения мы выбрали логистическую регрессию, рандомный лес и катбуст
    - Для первых двух моделей сделали кодирование категориальных признаков OHE и масштабирование RobustScaler, поскольку он не чувствителен к выбросам, которые у нас были в TotalCharges.
    - Для катбуста предобработка не нужна.
    
5. **Обучили три модели с перебором параметров в каждой и выбрали лучшую**:
    - Написали словарь параметров для каждой модели. Катбусту сделали отдельный словарь, поскольку он работает с признаками типа `category`, а остальные модели – с `object`, закодированными через OHE.
    - Обучили модели через RandomizedSearch

### Результаты

По итогам обучения лучшей моделью оказалась CatBoost с learning_rate = 0.1 и bootstrap_type = Bernoulli.

Важность признаков (SHAP) показала, что самыми важными признаками оказались количество дней, тип платежа, размеры ежемесячных и total платежей, а также наличие партнера. В целом, то же самое можно было увидеть и на корреляционном анализе. Остальные признаки тоже нашли свое место в модели.

Ее метрики:
- roc-auc на кросс-валидации: 0.88
- roc-auc на тестовой выборке: 0.9
- accuracy score на тестовой: 0.91


Модель делает успешные предсказания в 91% случаев. Но мы знаем, что у нас дисбаланс классов, поэтому сделали матрицу ошибок, чтобы посмотреть на предсказания детальнее:

- True Positive: модель хорошо предсказывает, кто останется (83%)
- True Negative: неплохо ловит ушедших (8.1%)
- False Positive: пропускает почти половину ушедших (7.6%)
- False Negative: мало (1.4%)

###  Рекомендации заказчику
Поскольку задача состояла в том, чтобы создать модель, которая будет предсказывать, разорвёт ли абонент договор, мы не знаем, что важнее – ошибки первого рода (считали, что клиент уйдёт, а он остался) или второго рода (пропустили реально ушедшего клиента).

В виде, в котором модель существует сейчас, мы пропускаем половину клиентов с риском ухода. Если компании критически важно снизить ошибку второго рода, в дальнейшем нужно снижать порог вероятности для предсказания класса 1 (договор прекратится).
