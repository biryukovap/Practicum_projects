# Анализ успешности видеоигр для интернет-магазина "██████"

## Введение

### Цель работы
Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

### План работ
1. Подготовка данных:
    - считывание;
    - обработка пропусков и дубликатов;
    - форматирование;
    - создание дополнительных колонок.
    
    
2. Исследовательский анализ:
    - анализ динамики выпуска игр в разрезе по годам;
    - определение жизненого цикла платформ на примере самых успешных;
    - определение актуального периода для дальнейшей работы;
    - выбор потенциально прибыльных платформ;
    - анализ глобальных продаж игр в разбивке по платформам;
    - оценка влияния отзывов пользователей и критиков на продажи по платформам;
    - анализ распределения игр по жанрам и оценка прибыльности.
    
    
3. Проверка гипотез:
     - средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
     - средние пользовательские рейтинги жанров Action и Sports разные.


4. Написание выводов.

### Описание исходных данных

- `Name` — название игры
- `Platform` — платформа
- `Year_of_Release` — год выпуска
- `Genre` — жанр игры
- `NA_sales` — продажи в Северной Америке (миллионы проданных копий)
- `EU_sales` — продажи в Европе (миллионы проданных копий)
- `JP_sales` — продажи в Японии (миллионы проданных копий)
- `Other_sales` — продажи в других странах (миллионы проданных копий)
- `Critic_Score` — оценка критиков (максимум 100)
- `User_Score` — оценка пользователей (максимум 10)
- `Rating` — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

Период: 1980–2016 гг. Данные за 2016 год могут быть неполными.

```python
# подключение библиотек
import pandas as pd
import numpy as np
import pandas as pd
from scipy import stats as st
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

### Шаг 1. Откроем файл с данными и изучим общую информацию

#### 1.1. Считываем данные

```python
try:
    games = pd.read_csv('games.csv', sep=',')
except:
    games = pd.read_csv('████████████', sep=',')
```

#### 1.2 Выводим информацию о датасете – первые строки и инфо

```python
display(games.head())
games.info()
```

На первый взгляд – некорректные названия столбцов, большое количество пропусков в последних трех колонках и несколько не самых оптимальных форматов. Разберемся с этим на следующем шаге.

### Шаг 2. Подготовим данные

#### 2.1. Заменим названия столбцов (приведем к нижнему регистру)

```python
games.columns = games.columns.str.lower()
games.columns
```

#### 2.2. Обработаем пропуски и дубликаты

##### 2.2.1. Переведем в нижний регистр все переменные в формате str, чтобы было удобнее избавляться от дубликатов, если они есть

```python
for row in games:
    try:
        games[row] = games[row].str.lower()
    except:
        pass
```

##### 2.2.2. Обработаем дубликаты

```python
#ищем дубликаты
print('Дубликаты:', games.duplicated().sum(), sep='\n')
```

Дубликатов нет, идем дальше.

##### 2.2.3. Посмотрим на пропуски

```python
#сразу посчитаем долю пропусков, чтобы понимать масштаб работ
(games.isna().mean()*100).sort_values(ascending=False)
```

Строки с пропусками в колонках `name`, `genre`, `year_of_release` можно удалить – это скорее всего ошибка парсинга и потеря наблюдений не велика. 

```python
games = games.dropna(subset=['name', 'genre', 'year_of_release']).reset_index(drop=True)
```

В колонках с отзывами и рейтингом также может быть ошибка парсинга или отсутствие данных вовсе, но там больше 40% пропущенных значений, поэтому оставим все как есть.\
\
Посмотрим на уникальные значения в колонке `user_score` – сейчас у нее формат `object`, хотя должен быть числовой.

```python
games['user_score'].sort_values().unique()
```

Да, там фигурирует значение `tbd` (вероятно, это 'to be determined', но составители базы так и не заполнили эти значения). Для дальнейшей работы поменяем их на NaN, потому что в других колонках могут быть ценные данные.

```python
games.loc[games['user_score'] == 'tbd', 'user_score'] = np.nan
```

#### 2.3. Преобразуем данные в нужные типы

```python
display(games.dtypes)
```

В каких столбцах заменим тип данных и почему:
- в `year_of_release` вместо `float` поставим `int`, потому что год не может быть дробным
- в `user_score` вместо `object`, поставим `float`, потому что рейтинг – это числовое значение

```python
games['year_of_release'] = games['year_of_release'].astype('int')
games['user_score'] = games['user_score'].astype('float')
```

#### 2.4 Посчитаем суммарные продажи во всех регионах и запишием их в отдельный столбец

```python
#добавляем столбец с подсчетами
games['game_sales_total'] = games[['eu_sales', 'jp_sales', 'na_sales', 'other_sales']].sum(axis=1)

#проверяем, все ли на месте
games.head()
```

### Шаг 3. Проведем исследовательский анализ данных

#### 3.1. Посмотрим, сколько игр выпускалось в разные годы. Важны ли данные за все периоды?

```python
# подсчитываем количество игр по годам
games_per_year = games['year_of_release'].value_counts().sort_index()

# строим график
plt.figure(figsize=(15, 8))
plt.plot(games_per_year.index, games_per_year.values, marker='o', color='#4CAF50', linewidth=2)

# настраиваем заголовок и подписи осей
plt.title('Количество выпущенных игр по годам', fontsize=22, fontweight='bold')
plt.xlabel('Год выпуска', fontsize=18)
plt.ylabel('Количество игр', fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

По графику видно, что:
- плавный рост пошел в 90-х
- большой скачок – в 2002
- пик производства – 2008-2009
- резкий спад в 2012

### Шаг 5. Проверим гипотезы

Для обоих гипотез, проверяемых ниже:
- Нулевая гипотеза — принимаемое по умолчанию предположение о том, различия между замеряемыми величинами стастистически не значимы.
- Альтернативная гипотеза – двусторонняя, то есть статистически важные различия есть и не имеет значения, в большую или меньшую сторону.
- Для проверки гипотез применили Т-критерий Стьюдента для двух независимых выборок, поскольку выборки у нас действительно независимы, а распределение Стьюдента можно применять для выборок любого размера (при большом количестве наблюдений оно приближается к нормальному).

#### 5.1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

Сформулируем гипотезы для этого вопроса:
- H0: Средние пользовательские рейтинги Xbox One **равны** пользовательским рейтингам PC.
- H1: Средние пользовательские рейтинги Xbox One **не равны** пользовательским рейтингам PC.

*Код ниже будет проверять двустороннюю гипотезу о том, что средние пользовательские рейтинги платформ Xbox One и PC отличаются.*

```python
print('Средние пользовательские рейтинги Xbox One:',
      actual_games.query('platform == "xone" and not (user_score.isna())')['user_score'].mean().round(2))
print('Средние пользовательские рейтинги PC:',
      actual_games.query('platform == "pc" and not (user_score.isna())')['user_score'].mean().round(2))

#устанавливаем такое пороговое значение, потому что наблюдений не много
#если бы были тысячи строк, снизили до 0.01 или даже меньше
alpha = 0.05

results = st.ttest_ind(actual_games.query('platform == "xone" and not (user_score.isna())')['user_score'],
                       actual_games.query('platform == "pc" and not (user_score.isna())')['user_score'])

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**Вывод**: различия в средних пользовательских рейтингах игр для Xbox One и PC статистически не значимы.

#### 5.2. Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Сформулируем гипотезы для этого вопроса:
- H0: Средние пользовательские рейтинги жанра Action **равны** пользовательским рейтингам жанра Sports.
- H1: Средние пользовательские рейтинги жанра Action **не равны** пользовательским рейтингам жанра Sports.

*Код ниже будет проверять двустороннюю гипотезу о том, что средние пользовательские рейтинги жанров Action и Sports отличаются.*

```python
print('Средние пользовательские рейтинги жанра Action:',
      actual_games.query('genre == "action" and not (user_score.isna())')['user_score'].mean().round(2))
print('Средние пользовательские рейтинги жанра Sports:',
      actual_games.query('genre == "sports" and not (user_score.isna())')['user_score'].mean().round(2))

#здесь наблюдений больше, поэтому можем снизить пороговое значение
alpha = 0.01

results = st.ttest_ind(actual_games.query('genre == "action" and not (user_score.isna())')['user_score'],
                       actual_games.query('genre == "sports" and not (user_score.isna())')['user_score'])

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**Вывод**: различия в средних пользовательских рейтингах игр жанров Action и Sports есть и они статистически значимы.

    - плавный рост пошел в 90-х;
    - большой скачок – в 2002;
    - пик производства – 2008-2009;
    - резкий спад в 2012.
    
    
2. Изучили жизненный цикл платформ на примере самых успешных: ps2, x360, ps3, wii, ds и ps. В среднем он составляет 11 лет.


3. Выбрали актуальный для исследования период: с 2014 по 2016 г. по трем причинам:
    - как раз в рамках последних 3 лет сможем рассмотреть, производство игр для каких платформ уже на излете, а для каких набирает обороты;
    - в 2013 году вышли две "сильные" консоли – ps4 и xone, а в 2014 уже утвердились на рынке.
    
    
4. Выяснили, что ставку в продажах следует сделать на игры для ps4 и xone - они на взлете. Также можно продавать игры на платформы, находящиеся в начале или середине своего "жизненного цикла": 3ds, psv, wiiu. И на ПК – на них спрос небольшой, но относительно стабильный.


5. Проанализировали продажи игр для каждой из избранных платформ: успешнее всего оказались продукты для ps4, wiiu и xone.
    
    
6. Посмотрели, как влияют оценки критиков и пользователей на продажи игр:
    - от оценок критиков положительно умеренно зависят продажи для платформ ps4, xone, 3ds и wiiu и слабо – pc и psv;
    - от оценок пользователей положительно умеренно зависят продажи для wiiu и слабо –  3ds.

     Поэтому справедливым будет сказать, что при покупке игроки больше опираются на оценки критиков, чем на пользовательские.
     
     
7. Посмотрели на общее распределение игр по жанрам:
    - самый прибыльный жанр – shooter;
    - sports и role-playing делят третье место.
    - жанры с самыми низкими продажами – puzzle, strategy и adventure.
    
    
8. Составили портрет пользователя каждого региона


    По платформам:
    - в Северной Америке доминирует продукция для Xbox (с почти пятой долей у стареющего Xbox 360), дальше идет портативная Wii U и PlayStation 4 и, наконец Wii – консоль с уникальными беспроводными контролерами;
    - в Европе первенство у продуктов PlayStation 4, дальше Wii, Xbox One, Wii U и следом Xbox 360;
    - Япония отдает предпочтение играм на Nintendo и PS, причем интересно, что в топ вошли две портативные консоли – psv и Wii U. Продукция Xbox в топ не входит.
    
    
    По жанрам:
    - Европа и Северная Америка больше всего играют в шутеры и спортивные игры. Также в обоих регионах популярны платформеры. Различия состоят в том, что в Северной Америке любят файтинги и ролевики, а в Европе – гонки и симуляторы.
    - Япония – уникальный регион, где почти треть всех продаж приходится на role-playing. Также там любят файтинги, платформеры и симуляторы. И пазлы – это самый не прибыльный жанр по миру, но, похоже, все продажи концентрируются на Японии.
    
    
    И рассмотрели влияние рейтинга на продажи:
    - В Европе и Северной Америке рейтинг влияет на продажи одинаково: больше всего продаются mature и everyone, затем e10+ и на последнем месте teen.
    - В Японии больше всего популярны игры категорий teen, everyone и mature, и только затем e10+.

**Результат проверки гипотез**:
1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые – это правда. Различия есть, но они статистически не значимы.


2. Средние пользовательские рейтинги жанров Action и Sports разные - и это правда. Даже если сравнивать средние вручную, разница будет в 1,5 балла – по 10-балльной шкале это чувствительная разница.
