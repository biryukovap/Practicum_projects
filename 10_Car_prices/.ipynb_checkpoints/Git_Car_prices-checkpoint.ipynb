# Определение стоимости автомобилей

## Задача

Сервис по продаже автомобилей с пробегом «████████████» разрабатывает приложение для привлечения новых клиентов. В нём можно быстро узнать рыночную стоимость своего автомобиля. В нашем распоряжении исторические данные: технические характеристики, комплектации и цены автомобилей. **Задача: построить модель для определения стоимости**. 

Заказчику важны:

- качество предсказания;
- скорость предсказания;
- время обучения.

## Описание данных

Данные находятся в файле `██████`.

Признаки:
- дата скачивания анкеты из базы
- тип автомобильного кузова
- год регистрации автомобиля

- пробег (км)
- месяц регистрации автомобиля

- марка автомобиля
- была машина в ремонте или нет

- количество фотографий автомобиля

- дата последней активности пользователя

Целевой признак:
- цена (евро)

```python
#функция для импорта данных
def data_import_os_request(data):
    pth1 = ( data + '.csv')
    pth2 = ('████████████' + data + '.csv')
    response = requests.get(pth2)
    
    if os.path.exists(pth1):
        df = pd.read_csv(pth1, sep=',', decimal=',')
        return df
    elif response.status_code==200:
        csv = BytesIO(response.content)
        df = pd.read_csv(csv, sep=',', decimal=',')
        return df
    else:
        print('Something is wrong')
```

### Деление на тренировочную и тестовую выборки

```python
X = ██████.drop('Price', axis=1)
y = ██████['Price']

X_train, X_test, y_train, y_test = \
train_test_split(X, y, test_size=0.25, random_state=RANDOM_STATE)
```

```python
# проверяем размеры выборок
display(X_train.shape)
display(y_train.shape)

display(X_test.shape)
display(y_test.shape)
```

### Подготовка данных

```python
#переводим object в category

def categorize(data):
    cat_columns = ['VehicleType', 'Gearbox', 'FuelType', 'Brand', 'Repaired']
    for col in cat_columns:
        data[col] = data[col].astype('category')
```

```python
categorize(X_train)
categorize(X_test)
```

```python
ohe_columns = ['VehicleType', 'FuelType', 'Brand']
ord_columns = ['Gearbox', 'Repaired']
num_columns = ['RegistrationYear', 'Power', 'Kilometer']
```

```python
# SimpleImputer + OHE
ohe_pipe = Pipeline(
    [
        (
            'simpleImputer_ohe', 
            SimpleImputer(missing_values=np.nan, strategy='constant', fill_value='unknown')
        ),
        (
            'ohe', 
            OneHotEncoder(handle_unknown='ignore', sparse_output=False)
        )
    ]
)

ord_pipe = Pipeline(
    [
        (
            'simpleImputer_ohe', 
            SimpleImputer(missing_values=np.nan, strategy='constant', fill_value='unknown')
        ),
        (
            'ord',
            OrdinalEncoder(
                categories=[
                    ['manual', 'auto'],
                    ['yes', 'no']
                ], 
                handle_unknown='use_encoded_value',
                unknown_value=np.nan)
        )
    ]
)

# создаем общий пайплайн предобработки
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns),
        ('ord', ord_pipe, ord_columns),
        ('num', StandardScaler(), num_columns)
    ], 
    remainder='passthrough'
)
```

```python
X_train_preprocessed = data_preprocessor.fit_transform(X_train)
X_test_preprocessed = data_preprocessor.transform(X_test)

ohe_feature_names = data_preprocessor.named_transformers_['ohe'].named_steps['ohe'].get_feature_names_out(input_features=ohe_columns)
all_feature_names = np.concatenate([ohe_feature_names, ord_columns, num_columns])
X_train_preprocessed = pd.DataFrame(X_train_preprocessed, columns=all_feature_names)
X_test_preprocessed = pd.DataFrame(X_test_preprocessed, columns=all_feature_names)
```

```python
#смотрим на итоги работы препроцессора
display(X_train_preprocessed.head(3))
display(X_test_preprocessed.head(3))
```

Данные предобработаны

## Обучение моделей

### Линейная регрессия

```python
# инициализируем модель линейной регрессии
model_lr = LinearRegression()

#прописываем параметры для перебора
param_grid_lr = {
    'fit_intercept': [True, False],
    'positive': [True, False]
}

# GridSearchCV
grid_search_lr = GridSearchCV(
    estimator=model_lr,
    param_grid=param_grid_lr,
    cv=5,
    scoring='neg_mean_squared_error',
    n_jobs=-1)

grid_search_lr.fit(X_train_preprocessed, y_train)
```

```python
lr_results = pd.DataFrame(grid_search_lr.cv_results_)
lr_results[['mean_fit_time', 'mean_score_time', 'params', 'mean_test_score', 'rank_test_score']].sort_values(by='rank_test_score')[:1]
```

```python
neg_mse_lr = (lr_results[['mean_fit_time', 'mean_score_time', 'params', 'mean_test_score', 'rank_test_score']]
              .sort_values(by='rank_test_score')[:1]['mean_test_score'])
rmse_lr = np.sqrt(abs(neg_mse_lr))
rmse_lr
```

###### Итоги линейной регрессии:
- Лучшие параметры: {'fit_intercept': True, 'positive': False}
- Время обучения: 1.241902	 секунд
- Время предсказания: 0.097144 секунд
- Значение RMSE: 2409 евро

### LightGBM

```python
param_lgb = {
            'feature_pre_filter': False,
            'learning_rate': 0.09,
            'max_depth': 9,
            'min_child_samples': 10,
            'subsample': 0.5,
            'reg_lambda': 5,
            'n_estimators': 1000,
            'force_col_wise': True,
            'verbose': -1
        }

model_lgb = lgb.LGBMRegressor(**param_lgb, random_state=RANDOM_STATE)
```

```python
%%time
model_lgb.fit(X_train,
              y_train,
              eval_metric='neg_mean_squared_error',
              categorical_feature=['VehicleType', 'Gearbox', 'FuelType', 'Brand', 'Repaired'])
```

```python
%%time
lgb_train_predictions = model_lgb.predict(X_train,
                  categorical_feature=['VehicleType', 'Gearbox', 'FuelType', 'Brand', 'Repaired'])
```

```python
rmse_lgb = np.sqrt(mean_squared_error(y_train, lgb_train_predictions))
rmse_lgb
```

- Параметры: {'feature_pre_filter': False,

### DecisionTreeRegressor

```python
model_dt = DecisionTreeRegressor(random_state=RANDOM_STATE)

param_dt = {
    'max_depth': range(5, 15),
    'min_samples_leaf': range(10, 31),
    'max_features': range(3, 9)
}

random_search_dt = RandomizedSearchCV(
    model_dt,
    param_dt,
    cv=5,
    scoring='neg_mean_squared_error',
    n_jobs=-1,
    random_state=RANDOM_STATE)

random_search_dt.fit(X_train_preprocessed, y_train)
```

```python
dt_results = pd.DataFrame(random_search_dt.cv_results_)
display(dt_results[['mean_fit_time', 'mean_score_time', 'params', 'mean_test_score', 'rank_test_score']].sort_values(by='rank_test_score')[:1])
print()
display(random_search_dt.best_params_)
```

```python
neg_mse_dt = (dt_results[['mean_fit_time', 'mean_score_time', 'params', 'mean_test_score', 'rank_test_score']]
              .sort_values(by='rank_test_score')[:1]['mean_test_score'])
rmse_dt = np.sqrt(abs(neg_mse_dt))
rmse_dt
```
