**Описание задачи**

Заказчик — кредитный отдел банка. Нужно разобраться, влияет ли семейное положение и количество детей клиента на факт погашения кредита в срок. Входные данные от банка — статистика о платёжеспособности клиентов.
Результаты исследования будут учтены при построении модели кредитного скоринга — специальной системы, которая оценивает способность потенциального заёмщика вернуть кредит банку.

**Описание данных**
- `children` — количество детей в семье
- `days_employed` — общий трудовой стаж в днях
- `dob_years` — возраст клиента в годах
- `education` — уровень образования клиента
- `education_id` — идентификатор уровня образования
- `family_status` — семейное положение
- `family_status_id` — идентификатор семейного положения
- `gender` — пол клиента
- `income_type` — тип занятости
- `debt` — имел ли задолженность по возврату кредитов
- `total_income` — ежемесячный доход
- `purpose` — цель получения кредита

**Этапы исследования:**

- загрузка и обзор данных из файла ████████████
- предобработка данных (удаление пропусков, обработка аномалий и дубликатов, изменение типов, категоризация)
- ответы на вопросы исследования
- общий вывод

## Предобработка данных

### Удаление пропусков

```python
# выведем количество пропущенных значений для каждого столбца
data.isna().sum()
```

В двух столбцах есть пропущенные значения. Один из них — `days_employed`. Пропуски в нем обработаем на следующем этапе.

Другой столбец с пропущенными значениями — `total_income` — хранит данные о доходах. На сумму дохода сильнее всего влияет тип занятости, поэтому заполнить пропуски в этом столбце нужно *медианным значением по каждому типу из столбца `income_type`*. Например, у человека с типом занятости `сотрудник` пропуск в столбце `total_income` должен быть заполнен медианным доходом среди всех записей с тем же типом.

```python
for t in data['income_type'].unique():
    data.loc[(data['income_type'] == t) & (data['total_income'].isna()), 'total_income'] = \
    data.loc[(data['income_type'] == t), 'total_income'].median()
```

### Обработка аномальных значений

В данных могут встречаться артефакты (аномалии) — значения, которые не отражают действительность и появились по какой-то ошибке. Таким артефактом будет отрицательное количество дней трудового стажа в столбце `days_employed`. Для реальных данных это нормально. Обработаем значения в этом столбце: заменим все отрицательные значения положительными с помощью метода `abs()`.

```python
data['days_employed'] = data['days_employed'].abs()
```

Для каждого типа занятости выведем медианное значение трудового стажа `days_employed` в днях.

```python
data.groupby('income_type')['days_employed'].agg('median')
```

У двух типов (безработные и пенсионеры) получаются аномально большие значения. Исправить такие значения сложно, поэтому оставим их как есть. Тем более этот столбец не понадобится для исследования.

Выведем перечень уникальных значений столбца `children`.

```python
data['children'].unique()
```

В столбце `children` есть два аномальных значения. Удалим строки, в которых встречаются такие аномальные значения из датафрейма `data`.

```python
data = data[(data['children'] != -1) & (data['children'] != 20)]
```

Ещё раз выведем перечень уникальных значений столбца `children`, чтобы убедиться, что артефакты удалены.

```python
data['children'].unique()
```

Заполним пропуски в столбце `days_employed` медианными значениями по каждого типа занятости `income_type`.

```python
for t in data['income_type'].unique():
    data.loc[(data['income_type'] == t) & (data['days_employed'].isna()), 'days_employed'] = \
    data.loc[(data['income_type'] == t), 'days_employed'].median()
```

Убедимся, что все пропуски заполнены.

```python
data.isna().sum()
```

### Изменение типов данных

Заменим вещественный тип данных в столбце `total_income` на целочисленный с помощью метода `astype()`.

```python
data['total_income'] = data['total_income'].astype(int)
```

### Обработка дубликатов

Обработаем неявные дубликаты в столбце `education`. В этом столбце есть одни и те же значения, но записанные по-разному: с использованием заглавных и строчных букв. Приведем их к нижнему регистру и проверим остальные столбцы.

```python
data['education'] = data['education'].str.lower()
```

```python
for c in ['family_status', 'gender', 'income_type', 'purpose']:
    print(c, ':', data[c].unique())
```

Выведем на экран количество строк-дубликатов в данных. Если такие строки присутствуют, удалим их.

```python
data.duplicated().sum()
```

```python
data = data.drop_duplicates()
```

### Категоризация данных

На основании диапазонов, указанных ниже, создадим в датафрейме `data` столбец `total_income_category` с категориями:
- 0–30000 — `'E'`;
- 30001–50000 — `'D'`;
- 50001–200000 — `'C'`;
- 200001–1000000 — `'B'`;
- 1000001 и выше — `'A'`.

Используем собственную функцию с именем `categorize_income()` и метод `apply()`.

```python
def categorize_income(income):
    try:
        if 0 <= income <= 30000:
            return 'E'
        elif 30001 <= income <= 50000:
            return 'D'
        elif 50001 <= income <= 200000:
            return 'C'
        elif 200001 <= income <= 1000000:
            return 'B'
        elif income >= 1000001:
            return 'A'
    except:
        pass
```

```python
data['total_income_category'] = data['total_income'].apply(categorize_income)
```

Выведем на экран перечень уникальных целей взятия кредита из столбца `purpose`.

```python
data['purpose'].unique()
```

Создадим функцию, которая на основании данных из столбца `purpose` сформирует новый столбец `purpose_category`, в который войдут следующие категории:
- `'операции с автомобилем'`,
- `'операции с недвижимостью'`,
- `'проведение свадьбы'`,
- `'получение образования'`.

Например, если в столбце `purpose` находится подстрока `'на покупку автомобиля'`, то в столбце `purpose_category` должна появиться строка `'операции с автомобилем'`.

Используем собственную функцию с именем `categorize_purpose()` и метод `apply()`.

```python
def categorize_purpose(row):
    try:
        if 'автом' in row:
            return 'операции с автомобилем'
        elif 'жил' in row or 'недвиж' in row:
            return 'операции с недвижимостью'
        elif 'свад' in row:
            return 'проведение свадьбы'
        elif 'образов' in row:
            return 'получение образования'
    except:
        return 'нет категории'
```

```python
data['purpose_category'] = data['purpose'].apply(categorize_purpose)
```

## Исследуем данные и ответим на вопросы

### Есть ли зависимость между количеством детей и возвратом кредита в срок?

Для начала ответим на вопрос, сколько клиентов в среднем являются задолжниками:

```python
in_debt_total = data['debt'].sum() / data['debt'].count()
print(f'Общая доля должников: {in_debt_total:.0%}')
```

```python
data_pivot_children = data.pivot_table(index='children', values='debt', aggfunc=['count', 'sum'])
data_pivot_children.columns = ['всего кредитополучателей', 'всего должников']
data_pivot_children['% должников'] = data_pivot_children['всего должников'] / data_pivot_children['всего кредитополучателей'] * 100
data_pivot_children
```

Сгруппируем данные по самым многочисленным категориям – семьям с 0, 1 и 2+ детьми.

```python
data_pivot_children_groupped = data_pivot_children.loc[2:, ['всего кредитополучателей', 'всего должников']].sum()
data_pivot_children = data_pivot_children[0:2].append(data_pivot_children_groupped, ignore_index=True)
data_pivot_children[['% должников']] = data_pivot_children['всего должников'] / data_pivot_children['всего кредитополучателей'] * 100
data_pivot_children
```

После группировки стало видно, что с ростом количества детей заемщики реже закрывают кредиты вовремя.

### Есть ли зависимость между семейным положением и возвратом кредита в срок?

```python
#рассчитываем долю должников внутри каждой группы

data_pivot_family = data.pivot_table(index='family_status', values='debt', aggfunc=['count', 'sum'])
data_pivot_family.columns = ['всего кредитополучателей', 'всего должников']
data_pivot_family['% должников'] = data_pivot_family['всего должников'] / data_pivot_family['всего кредитополучателей'] * 100
data_pivot_family
```

**Вывод:** 
- Вдовствующие и разведенные клиенты идентичны по частоте выплаты кредитов, но должников среди них чуть меньше, чем по всему банку в среднем.
- Женатые клиенты погашают кредиты с той же частотой, что и по всему банку в среднем.
- Те, кто находятся в гражданском браке или не находятся в браке совсем, чуть чаще имеют задолженности.

### Есть ли зависимость между уровнем дохода и возвратом кредита в срок?

```python
#рассчитываем долю должников внутри каждой группы
   
data_pivot_income = data.pivot_table(index='total_income_category', values='debt', aggfunc=['count', 'sum'])
data_pivot_income.columns = ['всего кредитополучателей', 'всего должников']
data_pivot_income['% должников'] = data_pivot_income['всего должников'] / data_pivot_income['всего кредитополучателей'] * 100
data_pivot_income
```

**Вывод**

Для выявления зависимости достаточное количество наблюдений только у заемщиков категорий B и C. Данные говорят о том, что заемщики категории C реже закрывают кредиты вовремя.

### Как разные цели кредита влияют на его возврат в срок?

```python
#рассчитываем долю должников внутри каждой группы

data_pivot_purpose = data.pivot_table(index='purpose_category', values='debt', aggfunc=['count', 'sum'])
data_pivot_purpose.columns = ['всего кредитополучателей', 'всего должников']
data_pivot_purpose['% должников'] = data_pivot_purpose['всего должников'] / data_pivot_purpose['всего кредитополучателей'] * 100
data_pivot_purpose
```

**Вывод:**
- Те, кто брал кредит на недвижимость, незначительно чаще погашали задолженность вовремя, чем в среднем по банку.
- Напротив, кредиты на операции с автомобилем и образование выплачивались в срок чуть реже.
- Займы на проведение свадьбы статистически не отличаются от среднего показателя по банку.

### Почему заполнить пропуски медианным значением — лучшее решение для количественных переменных?

Медиана лучше среднего, потому что она менее чувствительна к выбросам и экстремальным значениям. Например, если бы среди пенсионеров были бы несколько человек со сверхдоходами, а пропуски в данных мы бы заполняли средним, значения были бы искажены в большую сторону. С медианой такого не происходит.
