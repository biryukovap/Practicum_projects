### Описание задачи

Владелец молочного хозяйства ████████████ хочет купить бурёнок, чтобы расширить поголовье стада коров. Для этого он заключил выгодный контракт с ассоциацией пастбищ ████████████.\
\
Условия позволяют фермеру очень тщательно отобрать коров. Он определяет качество молока по строгой методике, и при этом ему нужно выполнять свой план развития молочного хозяйства. Фермер хочет, чтобы каждая бурёнка давала не менее 6000 килограммов молока в год, а её надой был вкусным — строго по его критериям, ничуть не хуже.\
\
**Задача**: разработать модель машинного обучения, которая поможет фермеру управлять рисками и принимать объективное решение о покупке. Нужно создать две прогнозные модели для отбора бурёнок в поголовье:
- Первая будет прогнозировать возможный удой коровы (целевой признак Удой);
- Вторая — рассчитывать вероятность получить вкусное молоко от коровы (целевой признак Вкус молока).
С помощью модели нужно отобрать коров по двум критериям:
- средний удой за год — не менее 6000 килограммов;
- молоко должно быть вкусным.

```python
# создаем константу RANDOM_STATE
RANDOM_STATE = 42

# выделяем в отдельные переменные целевой признак и входные признаки
X = ferma_model_1.drop(['milk'], axis=1)
y = ferma_model_1['milk']

# делим данные на тренировочные и тестовые, фиксируем random_state
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=RANDOM_STATE)


## подготовка признаков для масштабирования

# создаем переменную с экземпляром класса StandardScaler()
scaler = StandardScaler()

# обучаем и трансформируем количественные признаки из тренировочной выборки 
X_train_scaled = scaler.fit_transform(X_train[num_col_names])

# трансформируем количественные признаки в тестовой выборке
X_test_scaled = scaler.transform(X_test[num_col_names])

# создаем переменную с экземпляром класса OneHotEncoder()
# включаем параметр для избегания дамми-ошибок
# отключаем sparse-отображение
encoder = OneHotEncoder(drop='first', sparse_output=False)

# обучаем и трансформируем категориальные признаки из тренировочной выборки 
X_train_ohe = encoder.fit_transform(X_train[cat_col_names])

# выполняем преобразование категориальных признаков в тестовой выборке
X_test_ohe = encoder.transform(X_test[cat_col_names])

# получаем от кодировщика список новых столбцов и сохраняем в новой переменной 
encoder_col_names = encoder.get_feature_names_out()

# создаём тренировочный и тестовый датафреймы из закодированных и отмасштабированных данных
# для данных OHE названия столбцов берем из переменной encoder_col_names

X_train_ohe = pd.DataFrame(X_train_ohe, columns=encoder_col_names)
X_test_ohe = pd.DataFrame(X_test_ohe, columns=encoder_col_names)

X_train_scaled = pd.DataFrame(X_train_scaled, columns=num_col_names)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=num_col_names)

# объединяем все преобразованные данные в тренировочный и тестовый датафреймы
# сбрасываем индексы в датафрейме с числовыми индексами, чтобы избежать ошибок

X_train = pd.concat([X_train_ohe, X_train_scaled], axis=1)
X_test = pd.concat([X_test_ohe, X_test_scaled], axis=1)

# инициализируем модель линейной регрессии
model_lr = LinearRegression()

# обучаем модель на тренировочных данных
model_lr.fit(X_train, y_train)

# получаем предсказания модели на тестовых данных
# и сохраняем их в переменную predictions
predictions = model_lr.predict(X_test)
```

```python
# выделяем в отдельные переменные целевой признак и входные признаки
X = log_model.drop(['taste'], axis=1)
y = log_model['taste']

# делим данные на тренировочные и тестовые, фиксируем random_state
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=RANDOM_STATE)


## подготовка признаков для масштабирования

# создаем переменную с экземпляром класса StandardScaler()
scaler = StandardScaler()

# обучаем и трансформируем количественные признаки из тренировочной выборки 
X_train_scaled = scaler.fit_transform(X_train[num_col_names])

# трансформируем количественные признаки в тестовой выборке
X_test_scaled = scaler.transform(X_test[num_col_names])

# создаем переменную с экземпляром класса OneHotEncoder()
# включаем параметр для избегания дамми-ошибок
# отключаем sparse-отображение
encoder = OneHotEncoder(drop='first', sparse_output=False)

# обучаем и трансформируем категориальные признаки из тренировочной выборки 
X_train_ohe = encoder.fit_transform(X_train[cat_col_names])

# выполняем преобразование категориальных признаков в тестовой выборке
X_test_ohe = encoder.transform(X_test[cat_col_names])

# получаем от кодировщика список новых столбцов и сохраняем в новой переменной 
encoder_col_names = encoder.get_feature_names_out()

# создаём тренировочный и тестовый датафреймы из закодированных и отмасштабированных данных
# для данных OHE названия столбцов берем из переменной encoder_col_names

X_train_ohe = pd.DataFrame(X_train_ohe, columns=encoder_col_names)
X_test_ohe = pd.DataFrame(X_test_ohe, columns=encoder_col_names)

X_train_scaled = pd.DataFrame(X_train_scaled, columns=num_col_names)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=num_col_names)

# объединяем все преобразованные данные в тренировочный и тестовый датафреймы
# сбрасываем индексы в датафрейме с числовыми индексами, чтобы избежать ошибок

X_train = pd.concat([X_train_ohe, X_train_scaled], axis=1)
X_test = pd.concat([X_test_ohe, X_test_scaled], axis=1)

# инициализируем модель логистической регрессии
clf = LogisticRegression(random_state=RANDOM_STATE)

# обучаем модель на тренировочных данных
# и сохраняем их в переменную y_pred
clf = clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

#### Оцениваем качество модели с помощью метрик
- `accuracy` покажет долю верных ответов модели
- `precision` – не слишком ли часто модель выставляет класс 1 объектам класса 0
- `recall` – смогла ли модель классификации присвоить класс 1 всем объектам этого класса

```python
print('accuracy: ', accuracy_score(y_test, y_pred))
print('precision: ', precision_score(y_test, y_pred))
print('recall: ', recall_score(y_test, y_pred))
```

- Mодель ошибочно назвала 43 невкусных молока вкусными, а 12 вкусных пометила как невкусные. Нужно корректировать порог.

### Дорабатываем модель

```python
#выведем объекты по убыванию вероятности их принадлежности к классу 1.
y_proba = clf.predict_proba(X_test)[:,1]

data = pd.DataFrame(zip(y_test, y_proba),
             columns = ['y_test', 'y_proba']).sort_values(by='y_proba',ascending=False)

data.head()
```

```python
#сформируем возможные пороги
thresholds = [round(i,2) for i in np.linspace(0.1,1,num = 10, endpoint=False)]

thresholds
```

```python
# создаём переменную, которая будет хранить список новых колонок
columns = []

# добавим столбцы с новыми предсказаниями в таблицу
for i in thresholds:
    columns.append('y_pred_'+str(i))
    data['y_pred_'+str(i)] = data['y_proba'].apply(lambda y_proba: 1 if y_proba >=i else 0)

# выведем 5 случайных строк
data.sample(5)
```

```python
#напишем функцию, чтобы посмотреть метрики при разных порогах
def calculate_metrics(y_true, y_pred_proba, thresholds):
    results = []
 
    for threshold in thresholds:
        # Получаем предсказания для текущего порога
        y_pred = (y_pred_proba >= threshold).astype(int)
 
        # Строим матрицу ошибок
        tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
 
        # Вычисляем метрики
        recall = tp / (tp + fn) if (tp + fn) != 0 else 0
        precision = tp / (tp + fp) if (tp + fp) != 0 else 0
        true_positives = tp
        false_positives = fp
        false_negatives = fn
 
        # Сохраняем результаты для текущего порога
        results.append([f'y_pred_{threshold:.3f}', recall, precision, true_positives, false_positives, false_negatives])
 
    # Создаем DataFrame для итоговой таблицы
    columns = ['Порог', 'recall', 'precision', 'Количество "правильных единичек"', 'Количество False Positive', 'Количество False Negative']
    return pd.DataFrame(results, columns=columns)
 
metrics_table = calculate_metrics(data['y_test'], data['y_proba'], thresholds)
 
metrics_table
```

Порог ниже 0.73 дает много FP, чего нам как раз нужно избежать, но при этом очень много FN. Надо поискать пороги в интервале от 0.55 до 0.73.

```python
#сформируем возможные пороги в новом интервале
thresholds = [round(i,2) for i in np.linspace(0.55,0.82,num = 7, endpoint=False)]

# создаём переменную, которая будет хранить список новых колонок
columns = []

# добавим столбцы с новыми предсказаниями в таблицу
for i in thresholds:
    columns.append('y_pred_'+str(i))
    data['y_pred_'+str(i)] = data['y_proba'].apply(lambda y_proba: 1 if y_proba >=i else 0)

metrics_table = calculate_metrics(data['y_test'], data['y_proba'], thresholds)
 
metrics_table
```

В задаче не указано, сколько фермеру нужно купить коров, поэтому будем исходить из того, что нельзя сильно ошибиться со вкусом молока. Выберем порог в 0,7 – это сильно снизит риски для фермера.

```python
the_treshold = 0.7

y_pred_threshold = (y_proba > the_treshold).astype(int)

print('accuracy: ', accuracy_score(y_test, y_pred_threshold))
print('precision: ', precision_score(y_test, y_pred_threshold))
print('recall: ', recall_score(y_test, y_pred_threshold))

cm = confusion_matrix(y_test, y_pred_threshold)
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens_r')
plt.ylabel('Реальность')
plt.xlabel('Предсказание')
plt.title(f'Матрица ошибок', fontsize=18, fontweight='bold')
plt.show()
```

- Accuracy и recall снизились, потому что баланс между классами с новым порогом сместился в сторону невкусного молока
- Зато задача выполнена: подняли важную нам метрику precision
- Теперь модель ошибочно помечает невкусное молоко как вкусное всего в 7 случаях.
