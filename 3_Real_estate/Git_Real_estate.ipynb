### Предобработка данных

#### 1. Переименуем столбцы в соответствии со змеиным_языком

```python
data = data.rename(columns = {'cityCenters_nearest':'city_centers_nearest',
                              'parks_around3000':'parks_around_3000',
                              'ponds_around3000':'ponds_around_3000'})
```

#### 2. Изучим пропущенные значения в столбцах

```python
missing_data = pd.DataFrame({
    'Количество пропусков': data.isna().sum(),
    'Доля пропусков (%)': data.isna().mean() * 100
})

missing_data.sort_values(by='Количество пропусков', ascending=False)
```

**Общий вывод можно сделать такой**: пропуски в данных, которые заполняются пользователем, имеют человеческий фактор происхождения. А картографические параметры и данные о количестве дней могли возникнуть, например, из-за ошибки выгрузки или неправильно настроенных алгоритмов – в любом случае, стоит задать вопрос разработчикам.

Пропусков в `ceiling_height` много, поэтому заполним их медианным значением:

```python
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())
```

Строки с пропусками во `floors_total` можно удалить – это небольшое количество наблюдений.

```python
data = data.dropna(subset=['floors_total'])
```

Пропуски в `living_area` заполним, исходя из среднего соотношения жилой площади к общей – там высокий коэффициент корреляции.

```python
print('Коэффициент корреляции:', data['living_area'].corr(data['total_area']).round(1))
```

```python
data['typical_living_area'] = data['living_area'] / data['total_area']
data.loc[data['living_area'].isna(), 'living_area'] = data['total_area'] * data['typical_living_area'].median()

# и еще раз перезапишем столбец, чтобы не было пропусков
data['typical_living_area'] = data['living_area'] / data['total_area']
```

Пропуски в `is_apartment` и `balcony` можно заменить на `False` и `0` соответственно, потому что, скорее всего, при заполнении анкеты пользователь забыл указать, что балконов нет и квартира не является апартаментами.

```python
data['balcony'] = data['balcony'].fillna(0)
data['is_apartment'] = data['is_apartment'].fillna(False)
```

В студиях пропуски в `kitchen_area` можно заменить на `0`, потому что в студиях кухонь нет. В остальных случаях заполним, посчитав среднее значение отношения площади кухни ко всей площади.

```python
data.loc[data['studio'] == True, 'kitchen_area'] = data.loc[data['studio'] == True, 'kitchen_area'].fillna(0)

data['typical_kitchen_coeff'] = data['kitchen_area'] / data['total_area']
data['kitchen_area'] = data['kitchen_area'].fillna(data['total_area'] * data['typical_kitchen_coeff'].median())

# и еще раз перезапишем столбец, чтобы не было пропусков
data['typical_kitchen_coeff'] = data['kitchen_area'] / data['total_area']
```

Пропуски в `locality_name` можно заменить на `undefined`, так как логичную замену предположить нельзя:

```python
data['locality_name'] = data['locality_name'].fillna('undefined')
```

Строки с пропусками в `days_exposition` и картографических данных заменим на `-1`, так как логичную замену произвести невозможно, а `NaN` не позволит переводить значения в `int`.

```python
to_fill = ['days_exposition', 'airports_nearest', 'parks_around_3000',
     'parks_nearest', 'ponds_around_3000', 'ponds_nearest', 'city_centers_nearest']
def filler(data, col):
    data[col] = data[col].fillna(-1)
    
filler(data, to_fill)
```

Смотрим, что получилось:

```python
data.isna().sum()
```

```python
data.dtypes
```

Для упрощения работы с данными и оптимизации использования памяти форматы нужно поменять у столбцов:
- `last_price ` на `int`
- `first_day_exposition` на `datetime`
- `floors_total` на `int`
- `is_apartment` на `bool`
- `balcony` на `int`
- `locality_name` на `string`
- `days_exposition` на `int`
- `parks_around_3000` на `int`
- `ponds_around_3000` на `int`

```python
data['first_day_exposition']= pd.to_datetime(data['first_day_exposition'])
```

```python
change = ['last_price','floors_total', 'days_exposition', 'balcony', 'parks_around_3000', 'ponds_around_3000']
for _ in change:
    data[_] = data[_].astype('int')
    
data['is_apartment'] = data['is_apartment'].astype('bool')
data['locality_name'] = data['locality_name'].astype('string')
```

#### 4. Устраним неявные дубликаты в столбце с названиями

Предварительно вывели через `data['locality_name'].sort_values().unique()` все уникальные значения столбца, чтобы понять, как его очищать. Далее произведем саму очистку и оставим только названия населенных пунктов.

```python
data['locality_name'] = data['locality_name'].str.lower()

to_replace = ['городской поселок ', 'деревня ', 'коттеджный поселок ', 'поселок городского типа ',
              'поселок при железнодорожной станции ', 'поселок станции ', 'поселок ',
              'садоводческое некоммерческое товарищество ','садовое товарищество ', 'село ']

data['locality_name'] = (
    data['locality_name']
    .str.replace('ё', 'е')
    .replace(to_replace, '', regex=True)
) 

data['locality_name'].unique()
```

### Добавление новых признаков

#### 1. Цена одного квадратного метра (поделим стоимость объекта на его общую площадь, а затем округлим до двух знаков после запятой)

```python
data['m2_price'] = data['last_price'] / data['total_area'].round(2)
```

#### 2. День недели, месяц и год публикации объявления

```python
data['publication_day'] = data['first_day_exposition'].dt.weekday
data['publication_month'] = data['first_day_exposition'].dt.month
data['publication_year'] = data['first_day_exposition'].dt.year
```

#### 3. Tип этажа квартиры (значения — «первый», «последний», «другой»)

```python
def floor_type(row):
    if row['floor'] == 1:
        return 'первый'
    if row['floor'] == row['floors_total']:
        return 'последний'
    else:
        return 'другой'
    
data['floor_type'] = data.apply(floor_type, axis=1).astype('string')
```

#### 4. Расстояние до центра города в километрах (переведем из м в км и округлим до ближайших целых значений)

```python
data['city_center_km'] = (data['city_centers_nearest'] / 1000).round(0)
```

### Исследовательский анализ данных

### 1. Изучим данные прицельно

Для начала посмотрим, что распределение цены за квадратный метр нормальное:

```python
# функция для гистограммы
def do_hist(column, bins, range=(30000, 170000)):
    mean = data[column].mean()
    median = data[column].median()
    
    plt.figure(figsize=(8, 5))
    
    plt.hist(data[column], bins=bins, range=range, color='#4CAF50', edgecolor='black', alpha=0.8)
    plt.axvline(mean, color='blue', linestyle='--', linewidth=2, label=f'Среднее: {mean:.0f}')
    plt.axvline(median, color='orange', linestyle='--', linewidth=2, label=f'Медиана: {median:.0f}')

    plt.legend(fontsize=12, loc='upper right')
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
```

```python
do_hist('m2_price', 30)
plt.title(f'Распределение цены за кв.м.', fontsize=16, fontweight='bold')
plt.xlabel('Цена за кв.м.', fontsize=14)
plt.ylabel('Количество объектов', fontsize=14)
plt.show()
```

Распределение цен за кв.м. действительно нормальное. В среднем цена составляет около 100.000 руб.

Теперь отфильтруем данные по цене за квадратный метр и оставим только те наблюдения, которые попадают в +- 1.5 межквартильных размаха:

```python
q3, q1 = np.percentile(data['m2_price'],[75 ,25])
iqr = q3 - q1

median = data['m2_price'].median()

upper_boundary = median + 1.5 * iqr
lower_boundary = median - 1.5 * iqr

data = data[(lower_boundary < data['m2_price']) &  (data['m2_price'] < upper_boundary)]
```

#### 3.1. Общая площадь

```python
def do_scatter(column, title, metric):
    plt.figure(figsize=(10, 6))
    plt.scatter(data[column], data['last_price'], color='#4CAF50', alpha=0.08, edgecolor='black')

    plt.title(f'Зависимость конечной цены от {title}', fontsize=18, fontweight='bold')
    plt.xlabel(metric, fontsize=14)
    plt.ylabel('Цена', fontsize=14)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.grid(axis='both', linestyle='--', alpha=0.7)
```

    r>0,70≤1,00 – сильная положительная связь,
    r>-0,01≤-0,29 – слабая отрицательная связь,
    r>-0,30≤-0,69 – умеренная отрицательная связь,

Корреляция высокая.
