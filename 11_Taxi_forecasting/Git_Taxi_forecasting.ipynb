#  Прогнозирование заказов такси

Компания «████████» собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час.


## Задача
Построить модель для предсказания количества заказов такси на следующий час.

Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

## Шаги
1. Загрузить данные и выполнить их ресемплирование по одному часу.
2. Проанализировать данные.
3. Обучить разные модели с различными гиперпараметрами. Сделать тестовую выборку размером 10% от исходных данных.
4. Проверить данные на тестовой выборке и сделать выводы.

## Описание данных
Данные лежат в файле `████████`. Количество заказов находится в столбце `num_orders` (от англ. *number of orders*, «число заказов»).

```python
#работа с датафреймами и визуализацией
import os
import requests
from io import BytesIO
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.stattools import adfuller

pd.options.display.max_colwidth = 0

#классы для подготовки и оценки
from statsmodels.tsa.stattools import acf
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import TimeSeriesSplit
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error

#нужные модели
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from catboost import CatBoostRegressor
```

### Загрузка данных

```python
#функция для импорта данных
def data_import_os_request(data):
    pth1 = ( data + '.csv')
    pth2 = ('████████' + data + '.csv')
    response = requests.get(pth2)
    
    if os.path.exists(pth1):
        df = pd.read_csv(pth1, sep=',', decimal=',', parse_dates=[0])
        return df
    elif response.status_code==200:
        csv = BytesIO(response.content)
        df = pd.read_csv(csv, sep=',', decimal=',', parse_dates=[0])
        return df
    else:
        print('Something is wrong')
```

```python
#импортируем
data = data_import_os_request('████')
```

```python
#отображаем информацию
display(data.head())
print()
display(data.describe(include='all', datetime_is_numeric=True))
print()
print('Данные последовательны?', '\n', data['datetime'].is_monotonic_increasing)
```

### Ресемплирование по одному часу

```python
def check_stationarity(timeseries):
    # Используем тест Дики-Фуллера для проверки стационарности
    result = adfuller(timeseries, autolag='AIC')
    p_value = result[1]
    print(f'Результат теста: {result[0]}')
    print(f'P-value: {p_value}')
    print('Стационарный ряд' if p_value < 0.05 else 'Нестационарный ряд')


check_stationarity(data['num_orders'])
```

- С каждым месяцем посуточное количество заказов увеличивалось
- В августе шумов заметно больше
- Одинаковый паттерн в количестве заказов в зависимости от дня недели: самые "горячие" дни – вторник и суббота, самые спокойные – понедельник и среда
- Одинаковый паттерн в количестве заказов в зависимости от часа: спад ночью, нарастание с 6 утра с пиками в райне 11 и 16 часов, дальше спад и пик в районе полуночи.
- Графики автокорреляции говорят, что есть почасовая зависимость (пик на лаге 1), посуточная (пик на лаге 24) и понедельная (пик на лаге 7 при ресемплинге по дням).
- Ряд стационарный, поэтому прогнозировать данные проще

## Обучение

```python
#сюда будем складывать результаты
learning_results = []
```

### Создание признаков

На этапе анализа мы увидели, что на сумму влияют час заказа, день недели и день месяца. Напишем функцию, которая создаст эти признаки.

```python
def make_features(data, rolling_mean_size):
    data['hour'] = data.index.hour
    data['dayofweek'] = data.index.dayofweek
    
    lags = [1, 24, 168] #лаг на час, на сутки и на неделю
    
    for lag in lags:
        data[f'lag_{lag}'] = data['num_orders'].shift(lag)

    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()



df = data.copy()
make_features(df, 168)
```

```python
df.head()
```

### Делим данные на выборки

```python
train, test = train_test_split(df, shuffle=False, test_size=0.1)
train = train.dropna()

X_train = train.drop('lag_168', axis=1)
y_train = train['lag_168']
X_test = test.drop('lag_168', axis=1)
y_test = test['lag_168']
```

### Линейная регрессия

```python
model_lr = LinearRegression()

param_grid_lr = {
        'fit_intercept': [True, False],
        'positive': [True, False]
}

tscv = TimeSeriesSplit(n_splits=2, test_size=730)

grid_search_lr = GridSearchCV(
    estimator=model_lr,
    param_grid=param_grid_lr,
    cv=tscv,
    scoring='neg_root_mean_squared_error',
    n_jobs=-1)

grid_search_lr.fit(X_train, y_train)
```

```python
grid_search_lr.best_params_
```

```python
print("RMSE обучающей выборки:", abs(grid_search_lr.best_score_))
```

```python
learning_results.append(['LinearRegression',
                         grid_search_lr.best_params_,
                         abs(grid_search_lr.best_score_)])
```

### DecisionTreeRegressor

```python
model_dt = DecisionTreeRegressor(random_state=42)

param_grid_dt = {
    'max_depth': [3, 5, 10, 15, None],
    'min_samples_split': [2, 5, 10, 20],
    'min_samples_leaf': [1, 5, 10, 20],
    'max_features': [None, 'sqrt', 'log2'],
    'max_leaf_nodes': [10, 20, 50, None],
    'criterion': ['squared_error', 'friedman_mse']
}

tscv = TimeSeriesSplit(n_splits=2, test_size=730)

grid_search_dt = GridSearchCV(estimator = model_dt,
                              param_grid = param_grid_dt,
                              cv=tscv,
                              scoring='neg_root_mean_squared_error')

grid_search_dt.fit(X_train, y_train)
```

```python
print(f"Best parameters: {grid_search_dt.best_params_}")
```

```python
print("RMSE обучающей выборки:", abs(grid_search_dt.best_score_))
```

```python
learning_results.append(['DecisionTreeRegressor',
                         grid_search_dt.best_params_,
                         abs(grid_search_dt.best_score_)])
```

### CatBoost

```python
param_cb = {
    'iterations': [500, 1000],
    'learning_rate': [0.05, 0.1],
    'depth': range(2, 8)
}

model_cb = CatBoostRegressor(loss_function='RMSE', verbose=0)

tscv = TimeSeriesSplit(n_splits=2, test_size=730)

grid_search_cb = GridSearchCV(model_cb,
                                      param_cb,
                                      cv=tscv,
                                      scoring='neg_root_mean_squared_error')

grid_search_cb.fit(X_train, y_train)
```

```python
print(f"Best parameters: {grid_search_cb.best_params_}")
```

```python
print("RMSE обучающей выборки:", abs(grid_search_cb.best_score_))
```

```python
learning_results.append(['CatBoost',
                         grid_search_cb.best_params_,
                         abs(grid_search_cb.best_score_)])
```

### Итоги обучения

```python
learning_results = pd.DataFrame(learning_results, columns=['model', 'params', 'rmse_train'])
learning_results.sort_values(by='rmse_train')
```

На тренировочных данных лучше всего себя показала CatBoost, хотя все средние метрики моделей очень близки друг к другу.

## Тестирование лучшей модели

Используем на тестовых данных модель CatBoost и посмотрим на итоговую метрику.

```python
predictions_cb_test = grid_search_cb.best_estimator_.predict(X_test)
rmse_cb_test = np.sqrt(mean_squared_error(y_test, predictions_cb_test))

print('RMSE лучшей модели на тестовой выборке:', rmse_cb_test)
```

Вероятно, расхождение метрик при кросс-валидации и на тестовой выборке обусловлено тем, что в тестовую выборку попали августовские данные, где шумов больше, чем в предыдущие месяцы.

```python
results = pd.DataFrame(y_test)
results['predictions'] = predictions_cb_test
```

```python
results.plot(figsize=(13,8))
plt.title('Визуализация реальных и предсказанных значений', fontsize=20, fontweight='bold', color="#333333")
plt.xlabel('Дата', fontsize=16)
plt.ylabel('Количество заказов', fontsize=16)

plt.grid(True, linestyle='--', alpha=0.6)
sns.despine()
plt.legend(title="Тип данных", title_fontsize=16, fontsize=12, frameon=True, shadow=True, facecolor='white', loc='upper right')
plt.show()
```
